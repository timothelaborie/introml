```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler
pd.set_option('display.max_columns', None)
```

```python
# for file in ["train", "test"]:
file = "train"

df = pd.read_csv(file+"_features.csv")
df = df.sort_values(by=["pid"])

count = pd.DataFrame(df.groupby(["pid"]).count()).iloc[:,2:]
count.columns = [x+"_count" for x in count.columns]
count = count.fillna(0)
count.reset_index(drop=True, inplace=True)


count
```

```python

#which values are missing
columns = ["isnan_"+str(x) for x in range(1,409)]
data = np.array(df)[:, 3:]
data = data.reshape(data.shape[0]//12,34*12)
missingval = pd.DataFrame(data, columns=columns)
missingval = missingval.fillna(-9000)
missingval = (missingval==-9000).astype(int)

#total number of nan values
total_nan = missingval.sum(axis=1)
total_nan = pd.DataFrame(total_nan, columns=["total_nan"])

df = pd.read_csv(file+"_features.csv",index_col="pid")
df = df.sort_values(by=["pid"])
#forward fill missing values
df = df.groupby(["pid"]).ffill().reset_index()
df.set_index("pid",inplace=True)
#backward fill missing values
df = df.groupby(["pid"]).bfill().reset_index()

#change from first to last
derivative = pd.DataFrame(df.groupby(["pid"]).last()-df.groupby(["pid"]).first()).iloc[:,2:]
derivative.columns = [x+"_derivative" for x in derivative.columns]
derivative.reset_index(drop=True, inplace=True)
derivative = derivative.fillna(0)

#std
std = pd.DataFrame(df.groupby(["pid"]).std()).iloc[:,2:]
std.columns = [x+"_std" for x in std.columns]
std.reset_index(drop=True, inplace=True)
std = std.fillna(0)

# #min
# min = pd.DataFrame(df.groupby(["pid"]).min()).iloc[:,2:]
# min.columns = [x+"_min" for x in min.columns]
# min.reset_index(drop=True, inplace=True)
# min = min.fillna(0)

# #max
# max = pd.DataFrame(df.groupby(["pid"]).max()).iloc[:,2:]
# max.columns = [x+"_max" for x in max.columns]
# max.reset_index(drop=True, inplace=True)
# max = max.fillna(0)
```

```python
derivative
```

```python



#group every 12 rows together
data = np.array(df)[:, 3:]
data = data.reshape(data.shape[0]//12,34*12)
# data = pd.DataFrame(data, columns=np.concatenate([[col+"_"+str(x) for x in range(1,13)] for col in df.columns[3:]]))
data = pd.DataFrame(data, columns=np.concatenate([[col+"_"+str(x) for col in df.columns[3:]] for x in range(1,13)]))

#append
df1 = df.iloc[::12,:3]
df2 = data
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)
```

```python
count
```

```python
df = pd.concat([df1,df2,derivative,count,std,total_nan],axis=1)
```

```python
df
```

```python



#replace inf with nan
df = df.replace([np.inf, -np.inf], np.nan)

#sort columns, but with pid, time age and total_nan at the front
df = df.reindex(sorted(df.columns), axis=1)
df = df[["pid","Time","Age","total_nan"]+[col for col in df.columns if col not in ["pid","Time","Age","total_nan"]]]

#fill nan
df = df.fillna(-9000)

#sort by pid
df = df.sort_values(by=["pid"])


# df = df[["pid"] + [col for col in df.columns if col.__contains__("Creatinine")]]
df.head()
```
