```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import pickle
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(2, 5, figsize=(20,10))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
        
    plt.tight_layout(h_pad=1, w_pad=1)
    plt.show()
```

```python
features = {}
# load the features dictionary from the file
with open('features_xl.pickle', 'rb') as handle:
    features = pickle.load(handle)
```

```python
df = pd.DataFrame(features)
df = df.T
df = df.sort_values(by=19, ascending=False)
# df
top = df.head(10)
top = np.array(top.index)
```

```python
imgs = [mpimg.imread("food_512/" + str(index) + ".jpg") for index in top]
plotImages(imgs)
```
